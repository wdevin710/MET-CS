```python
import csv
```

```python
List1=[]
```

```python
#Question 1
with open('/Users/haowu/Desktop/Boston University Graduate Study/CS 677/assignment 1/diamonds.csv','r') as file:
    diamonds = csv.reader(file, delimiter=',')
    for lines in diamonds:
        value = lines[2]
        if value == "Ideal":
            List1.append(lines)
print(List1)
```

```python
print("Total entries are:",len(List1))
```

```python
weight = 0
for i in List1:
    weight += float(i[1])
print("Average weight for Ideal group is:",weight/len(List1))
```

```python
price = 0
for i in List1:
    price += float(i[7])
print("Average price for Ideal group is:", price/len(List1))
```

```python
#Question 2
#Method a
method1 = 0
for i in List1:
    method1 += (float(i[7])/float(i[1]))
print("Average prices per carat for method A is:", method1/len(List1))
```

```python
#Method b
weightb = 0
priceb = 0
for i in List1:
    weightb += float(i[1])
    priceb += float(i[7])
print("Average prices per carat for method B is:",priceb/weightb)
print("Method A average price is lower")
```

```python
mxprice = None
for i in List1:
    j = (float(i[7])/float(i[1]))
    if(mxprice is None or j > mxprice) :
        mxprice = j
print("Maximum price per carat is:",mxprice)
```

```python
miprice = None
for i in List1:
    k = (float(i[7])/float(i[1]))
    if(miprice is None or k < miprice):
        miprice = k
print("Minimun price per carat is:",miprice)
```

```python
List2 = []
for i in List1:
    l = float(i[7])/float(i[1])
    List2.append(l)
List2.sort()
n = int(len(List2)/ 2)
print("Median price per carat is:",List2[n])
```

```python
#Question 3
#we fixed color of diamonds as "D" and Clarily as "IF" in group "Ideal"
List3 = []
for i in List1:
    color = i[3]
    clarily = i[4]
    if(color == "D" and clarily == "IF"):
        List3.append(i)
q3a = 0
for p in List3:
    q3a += (float(p[7])/float(p[1]))
print("Question 3 method a:", q3a/len(List3))
q3price = 0
q3weight = 0
for p in List3:
    q3price += float(p[7])
    q3weight += float(p[1])
print("Question 3 method b:", q3price/q3weight)
```
